#### Features

Descriptions of each feature in the dataset.
Refer to the [data codebook](https://www.cdc.gov/brfss/annual_data/2023/zip/codebook23_llcp-v2-508.zip) for the corresponding value definitions.

- **_STATE**: In FIPS (Federal Information Processing Standards code) Code. US States and protectorates number code.
- **GENHLTH**: General health status. (1=Excellent, 2=Very Good, etc.)
- **PHYSHLTH**: The number of days in the past 30 days when physical health was not good. 
- **MENTHLTH**: The number of days in the past 30 days when mental health was not good.
- **POORHLTH**: The number of days in the past 30 days during which poor physical or mental health interfered with usual activities, such as self-care, work, or recreation.
- **MEDCOST1**: Indicates whether, at any point in the past 12 months, there was a need to see a doctor but the individual was unable to do so due to financial constraints.
- **CHECKUP1**: The time elapsed since the individual last visited a doctor for a routine checkup, in years.
- **EXEROFT1**: The frequency, measured in times per week or per month, of participation in this activity during the past month.
- **BPHIGH6**: Indicates whether a doctor, nurse, or other health professional has ever diagnosed the individual with high blood pressure. For female respondents, clarifies if the diagnosis was only during pregnancy.
- **BPMEDS1**: Indicates whether the individual is currently taking prescription medication to manage high blood pressure. 
- **CHOLCHK3**: The time elapsed since the individual last had their cholesterol checked, in years.
- **CHOLMED3**: Indicates whether the individual is currently taking medication prescribed by a doctor or health professional to manage cholesterol levels. 
- **CVDINFR4**: Indicates whether the individual has ever been diagnosed with a heart attack, also known as a myocardial infarction.
- **CVDCRHD4**: Indicates whether the individual has ever been diagnosed with angina or coronary heart disease.
- **CVDSTRK3**: Indicates whether the individual has ever been diagnosed with a stroke.
- **ASTHNOW**: Indicates whether the individual currently has asthma.
- **CHCSCNC1**: Indicates whether the individual has ever been diagnosed with skin cancer that is not melanoma.
- **CHCOCNC1**: Indicates whether the individual has ever been diagnosed with melanoma or any other types of cancer.
- **CHCCOPD3**: Indicates whether the individual has ever been diagnosed with Chronic Obstructive Pulmonary Disease (COPD), emphysema, or chronic bronchitis.
- **ADDEPEV3**: Indicates whether the individual has ever been diagnosed with a depressive disorder, including depression, major depression, dysthymia, or minor depression.
- **CHCKDNY2**: Indicates whether the individual has ever been diagnosed with kidney disease.
- **HAVARTH4**: Indicates whether the individual has ever been diagnosed with arthritis or related conditions.
- **DIABETE4**: Indicates whether the individual has ever been diagnosed with diabetes, specifying if it was only during pregnancy for females or if it was pre-diabetes/borderline diabetes.
- **MARITAL**: Marital Status
- **EDUCA**: Education Level
- **VETERAN3**: Indicates whether the individual has ever served on active duty in the United States Armed Forces.
- **EMPLOY1**: Employment Status
- **INCOME3**: Income Level
- **PREGNANT**: Pregnancy Status
- **DECIDE**: Indicates whether the individual has serious difficulty concentrating, remembering, or making decisions due to a physical, mental, or emotional condition.
- **DIFFWALK**: Indicates whether the individual has serious difficulty walking or climbing stairs.
- **DIFFDRES**: Indicates whether the individual has difficulty dressing or bathing.
- **DIFFALON**: Difficulty Doing Errands Alone
- **SMOKDAY2**: Frequency of Days Now Smoking
- **USENOW3**: Use of Smokeless Tobacco Products
- **ALCDAY4**: Days in past 30 had alcoholic beverage
- **CIMEMLO1**: Indicates whether the individual has experienced worsening or more frequent difficulties with thinking or memory in the past 12 months.
- **CDWORRY**: Indicates whether the individual is concerned about their difficulties with thinking or memory.
- **CDDISCU1**: Indicates whether the individual or someone else has discussed their difficulties with thinking or memory with a healthcare provider.
- **CDHOUS1**: Indicates whether difficulties with thinking or memory have interfered with daily activities, such as managing medications, paying bills, or keeping appointments, in the past 12 months.
- **CDSOCIA1**: Indicates whether difficulties with thinking or memory have impacted the individual's ability to work or volunteer in the past 12 months.
- **FIREARM5**: Indicates whether any firearms are currently kept in or around the individual's home.
- **MARIJAN1**: The number of days in the past 30 days that the individual used marijuana or cannabis.
- **ACEDEPRS**: Indicates whether the individual lived with someone who was depressed, mentally ill, or suicidal.
- **LSATISFY**: Reflects the individual's overall level of satisfaction with their life.
- **EMTSUPRT**: Indicates the frequency with which the individual receives the social and emotional support they need.
- **SDLONELY**: Indicates the frequency with which the individual experiences feelings of loneliness.
- **SDHEMPLY**: Indicates whether the individual has experienced job loss or a reduction in work hours in the past 12 months.

- **SDHBILLS**: Indicates whether, in the past 12 months, there was a time when the individual was unable to pay their mortgage, rent, or utility bills.
- **SDHUTILS**: Indicates whether, in the past 12 months, an electric, gas, oil, or water company threatened to shut off services.

```python
def crosstab_feature_to_target(
    data,
    feature, 
    target,
    figsize=(10, 6), 
    legend_labels={'1.0':'0', '2.0':'1-13 Days', '3.0':'14+ Days', '9.0':'Don`t know'}, 
    feature_labels=[],
    legend_loc='upper right'
):
    """
        Explore how a categorical feature relates to the target 
        using cross-tab and grouped bar chart.
    """
    plt.figure(figsize=figsize)
    # Cross-tabulation
    crosstab = pd.crosstab(data[feature], data[target], normalize='index')
    crosstab_pct = crosstab.div(crosstab.sum(axis=1), axis=0) * 100
    fmt_crosstab = crosstab_pct.map(lambda x: f"{x:.2f}%")
    
    # Plotting the grouped bar chart
    ax = sns.countplot(x=feature, hue=target, data=data, palette='muted')
    
    for container, hue_category in zip(ax.containers, crosstab.columns):
        for bar, percentage in zip(container, crosstab_pct[hue_category]):
            height = bar.get_height()
            ax.text(
                bar.get_x() + bar.get_width() / 2, 
                height, 
                f'{percentage:.2f}%', 
                ha='center', 
                va='bottom', 
                fontsize=8,
                rotation=35
            )

    if len(legend_labels):
        handles, labels = ax.get_legend_handles_labels()
        mapped_labels = [legend_labels[label] for label in labels]
        plt.legend(handles, mapped_labels, title=target, loc=legend_loc)
    else:
        plt.legend(title=target, loc='center')
        
    if len(feature_labels):
        current_xticklabels = ax.get_xticklabels()
        new_labels = [feature_labels.get(label.get_text(), label.get_text()) for label in current_xticklabels]
        plt.xticks(ticks=range(len(new_labels)), labels=new_labels, fontsize=10, rotation=45)
    else:
        plt.xticks(fontsize=10, rotation=45)

    plt.title(f'Statistics of Mental Stress (Days) for {feature}', fontsize=14)
    plt.xlabel(None)
    plt.ylabel('Count')
    _ = plt.tight_layout()
    plt.show()
```

#### Analyzing Data Distributions

Check for imbalance. Set the threshold to 0.8. (Use the strictest threshold to ensure better performance, as required for healthcare models.)

```python

imbalance_threshold = 0.8

def check_feature_imbalance(df):
    imbalance_report = {}

    for column in df.columns:
        if df[column].dtype == 'object' or df[column].dtype.name == 'category':
            # Categorical: Check frequency distribution
            value_counts = df[column].value_counts(normalize=True)
            max_percentage = value_counts.max()
            imbalance_report[column] = {
                "type": "categorical",
                "max_percentage": max_percentage,
                "is_imbalanced": max_percentage > imbalance_threshold,
                "distribution": value_counts.to_dict()
            }
        else:
            # Numerical: Check distribution
            skewness = df[column].skew()
            bin_counts, bin_edges = np.histogram(df[column], bins=10)
            imbalance_report[column] = {
                "type": "numeric",
                "skewness": skewness,
                "bin_counts": bin_counts.tolist(),
                "bin_edges": bin_edges.tolist(),
                "is_imbalanced": np.any(bin_counts > imbalance_threshold * len(df))
            }

    return pd.DataFrame.from_dict(imbalance_report, orient='index')

imbalance_report = check_feature_imbalance(mh_df)
```

```python
# Get features that are imbalanced based on the 80% threshold setting, 
# then group numeric, and categorical features in a separate list

imbalanced_features = [] 
target_feature = None

for col, report in imbalance_report.iterrows():
    if report['is_imbalanced'] == True:
        imbalanced_features.append([col, report])
    if col == target:
        target_feature = report

imb_categorical = [report for report in imbalanced_features if report[1]['type'] == 'categorical']
imb_numeric = [report for report in imbalanced_features if report[1]['type'] == 'numeric']

print(f"Number of imbalanced features: {len(imbalanced_features)}; categorical: {len(imb_categorical)}; numeric={len(imb_numeric)}")
```

##### Categorical features with class imbalance.

```python
import math

# Create subplot of categorical charts

charts_per_row = 4
cat_num_charts = len(imb_categorical)
# Categorical charts subplot
if cat_num_charts:
    num_rows = math.ceil(cat_num_charts / charts_per_row)
    cat_fig, cat_ax = plt.subplots(
        num_rows,
        charts_per_row, 
        figsize=(15, num_rows * 3), 
        gridspec_kw={'hspace': 0.6, 'wspace': 0.3}
    )
    cat_ax = cat_ax.flatten()

# Plotting Categorical Feature Imbalance
for i, (column, row) in enumerate(imb_categorical):
    ax = cat_ax[i]
    pd.Series(row['distribution']).plot(kind='bar', color='skyblue', ax=ax)
    ax.set_title(f"{column}: max-pct={row['max_percentage']:.3f}")
    ax.set_xlabel("Class")
    ax.set_ylabel("Frequency")

# Remove extra subplots
for j in range(len(imb_categorical), len(cat_ax)):
    cat_fig.delaxes(cat_ax[j])

plt.show()
```

##### 2. Using LIME as alternative for Feature Selection - and Feature Explanation

While the previous approach combined model-agnostic feature selection methods like KBest with model-based techniques, **LIME (Local Interpretable Model-agnostic Explanations)** provides an alternative by focusing on explaining individual predictions. It works with any model and helps identify the most important features in a simple, clear way.

Our purpose for the Feature Explanation is to understand how individual features influence the model's predictions, enhancing the transparency and reliability of our approach to assessing the initial state of mental health decline.

The analysis focuses on two key classes: Class2 (1-13 Days), which is more prominently represented in the dataset, and Class3 (14+ Days), which is less frequent. 

To generate meaningful feature explanations efficiently, we will use LIME, instead of SHAP, for its speed and practicality. 

Additionally, we will use a Random Forest model for feature importance analysis due to its ability to handle both non-linear and complex relationships between features. 

```python

Xr_train, xr_test = np.array(X_train), np.array(x_test)

# Train a Random Forest Classifier
model = RandomForestClassifier(random_state=42)
model.fit(Xr_train, y_train)

# Create a LIME Tabular Explainer
explainer = LimeTabularExplainer(
    Xr_train,
    feature_names=X_train.columns,
    class_names=['0 Days', '1-13 Days', '14+ Days', 'Not Sure'],
    discretize_continuous=True
)

# Pick a single instance for explanation
instance = xr_test[0].reshape(1, -1)

# Explain the prediction for the selected instance
explanation = explainer.explain_instance(
    instance[0],
    model.predict_proba,
    num_features=k_important_features
)
```

##### 3. Feature selection through modeling derived and original features 



Building on the insights from the correlation analysis, let's explore interaction terms among socioeconomic, behavioral, and health features to capture more complex relationships.

The focus here is on feature selection, to identify the important predictors: Combining related variables into a socioeconomic index using PCA; create interaction terms to capture potential relationships for factors like income and education.

This insight will guide further analysis and potential interventions.


```python
###### Group by continuous and categorical features

# Create a disposable copy for this analysis.
# - Drop MENTHLTH 
df = mh_df.drop(columns=['MENTHLTH']).copy()

behavioral_features = ['EXEROFT1', 'CHOLCHK3']
mental_health_features = ['EMTSUPRT', 'ADDEPEV3', 'POORHLTH']
income_education_features = ['INCOME3', 'EDUCA']

socioeconomic_features = ['FOODSTMP', 'SDHBILLS', 'SDHFOOD1'] # PCA candidates
continuous_features = ['PHYSHLTH', 'POORHLTH', 'MARIJAN1'] # Numeric
#exclusive_features = list(set(socioeconomic_features + continuous_features + behavioral_features + mental_health_features + income_education_features))
exclusive_features = socioeconomic_features + continuous_features

categorical_features = [col for col in df.columns if col not in (exclusive_features + [target])]
```

```python
###### Create interaction terms for:

from imblearn.over_sampling import SMOTENC
from imblearn.pipeline import Pipeline as ImbPipeline
from imblearn.under_sampling import RandomUnderSampler

import warnings
warnings.filterwarnings("ignore", category=FutureWarning, message=".*_check_n_features.*")
warnings.filterwarnings("ignore", category=FutureWarning, message=".*_check_feature_names.*")
warnings.filterwarnings("ignore", category=FutureWarning, message=".*force_all_finite.*")
warnings.filterwarnings("ignore", category=FutureWarning, message=".*__sklearn_tags__.*")

# 1. Socioeconomic variables using PCA for feature reduction.
encoder = OneHotEncoder(drop='first')
encoded = encoder.fit_transform(df[socioeconomic_features])
scaler = StandardScaler(with_mean=False)
scaled = scaler.fit_transform(encoded)
socioeconomic_pca = PCA(n_components=1)
df['Socioeconomic_Index'] = socioeconomic_pca.fit_transform(scaled)

# 2. Behavioral and Preventive Health Variables.
# Using Cross-product feature
df['Behavioral_cross'] = df['EXEROFT1'].astype(int) * df['CHOLCHK3'].astype(int)

# 3. Health Interdependencies (Polynomial Interactions)
# Using Nonlinear interaction
df['Physical_Mental_Interaction'] = df['GENHLTH'].astype(int) * df['PHYSHLTH'] #* df['MENTHLTH']

# 4. Income and Education Interaction
df['Income_Education_Interaction'] = df['INCOME3'].astype(int) * df['EDUCA'].astype(int)

# 5. Mental Health
df['Mental_Health_Composite'] = df[mental_health_features].mean(axis=1)

######  Encode categorical features and scale continuous and aggregated features

# Get predictors and target
#df, _ = train_test_split(df, train_size=0.3, stratify=df[target], random_state=42)
X = df.drop(columns=socioeconomic_features + [target])
y = df[target]

X_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y, random_state=42)

aggregated_features = [
  'Mental_Health_Composite',
  'Income_Education_Interaction', 
  'Physical_Mental_Interaction',
  'Behavioral_cross',
  'Socioeconomic_Index'
]

## Setup pre-procssing and modeling pipeline
preprocessor = ColumnTransformer([
    ('scale', StandardScaler(), continuous_features + aggregated_features),
    ('encode', OneHotEncoder(handle_unknown='ignore'), categorical_features)
], remainder='passthrough', force_int_remainder_cols=False)

smotenc = SMOTENC(
  categorical_features=[X.columns.get_loc(col) for col in categorical_features],
  random_state=42
)

model = RandomForestClassifier(
  class_weight='balanced',
  max_depth=3000,
  n_jobs=-1,
  random_state=42,
  verbose=0
)

pipeline = ImbPipeline(steps=[
  ('preprocessor', preprocessor),
  ('smotenc', smotenc),
  ('undersample', RandomUnderSampler(random_state=42)), # smotenc: Undersample overrepresented classes.
  ('classifier', model)
])

###### Prepare for model training and prediction

# Train the model
pipeline.fit(X_train, y_train)

# Test prediction
y_pred = pipeline.predict(x_test)
```

```python

###### Plotting Feature importance

#classifier = model.named_steps['classifier']
importances = model.feature_importances_

# Get feature names after preprocessing
#categorical_encoded_features = pipeline.named_steps['preprocessor'].named_transformers_['encode'].get_feature_names_out(categorical_features)
categorical_encoded_features = preprocessor.named_transformers_['encode'].get_feature_names_out(categorical_features)
feature_names = continuous_features + aggregated_features + list(categorical_encoded_features)

print(f'importances={len(importances)}')
print(f'features={len(feature_names)}')

# Get top ten most important features
indices = np.argsort(importances)[::-1][:10]

# Plot the resulting 10 most important features
plt.figure(figsize=(12, 6))
plt.title("Feature Importances")
#plt.bar(range(len(feature_names)), importances[indices])
plt.bar(range(len(indices)), importances[indices])
plt.xticks(range(len(indices)), [feature_names[i] for i in indices], rotation=90)
plt.tight_layout()
plt.show()
```
